```python
# Pandas library for reading csv files and displaying them
import pandas as pd      
# to implement knn algorithm
from sklearn.neighbors import KNeighborsClassifier
# seaborn library to visualize data
import seaborn as sns                                        
```

```python
data = pd.read_csv("iris.csv")     # reading csv files
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
data.head()
df = pd.DataFrame(data, columns=['sepal_length','sepal_width','petal_length','petal_width','species'])
# df    #displaying csv file data
```

```python
sns.pairplot(data, hue="species", palette='bright')  # Visulaizing Data using seaborn and pairplot
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']   # Variables which will be used to predict
x = data[features].values
y = data['species'].values
```

```python
from sklearn.model_selection import train_test_split         # to split data into test and training data

x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 0, test_size = .2)  #splitting data into training and testing data
```

```python
knn1 = KNeighborsClassifier(n_neighbors = 3)    #Implementing KNN Algorithm
```

```python
knn1.fit(x_train, y_train)       # Training model with training data

y_pred_1 = knn1.predict(x_test)    # testing model with test data
```

```python
from sklearn.metrics import accuracy_score                   # to find accuracy of model

print("Accuracy with k = 3: ", accuracy_score(y_test, y_pred_1)*100)   # To find accuracy of the model
```

```python
predict1 = knn1.predict([[6,3,6,2],])    # should give output as virginca
print(predict1)
predict2 = knn1.predict([[6,3,4,1]])     # should give output as versicolor
print(predict2)
predict3 = knn1.predict([[5,3,1,0]])     # should give output as setosa
print(predict3)
```
